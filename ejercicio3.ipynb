```python
import pandas as pd
import numpy as np
import os
```

```python
#carga el csv
consulta = pd.read_csv("archivos/cybersecurity_attacks.csv")
```

```python
#Consultamos las primeras filas
consulta.head(10)
```

```python
#Eliminanos los valores nulos
consulta = consulta.dropna()
```

```python
#Eliminamos valores duplicados
consulta = consulta.drop_duplicates()
```

```python
#Eliminar valores atípicos en "Source Port" usando rango intercuartil (IQR)
Q1 = consulta['Source Port'].quantile(0.25)
Q3 = consulta['Source Port'].quantile(0.75)
IQR = Q3 - Q1
consulta = consulta[(consulta['Source Port'] >= Q1 - 1.5 * IQR) & (consulta['Source Port'] <= Q3 + 1.5 * IQR)]
```
